```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# set-up data
num_locations = 6
Q = 12
demands = [2, 1, 3, 2, 1, 2]

distances = [
    [0, 7, 12, 15, 20, 25],
    [7, 0, 5, 10, 15, 18],
    [12, 5, 0, 6, 10, 13],
    [15, 10, 6, 0, 8, 11],
    [20, 15, 10, 8, 0, 6],
    [25, 18, 13, 11, 6, 0]
]

times = [
    [0, 20, 25, 15, 10, 20],
    [20, 0, 10, 20, 25, 30],
    [25, 10, 0, 15, 20, 25],
    [15, 20, 15, 0, 15, 20],
    [10, 25, 20, 15, 0, 10],
    [20, 30, 25, 20, 10, 0]
]
```

```python
# create the graph
G = nx.Graph()

# add nodes to the graph
for i in range(num_locations):
    G.add_node(i + 1)

# add edges to the graph with distances as edge weights
for i in range(num_locations):
    for j in range(i + 1, num_locations):
        distance = distances[i][j]
        G.add_edge(i + 1, j + 1, weight=distance)

# create a layout for the graph
pos = nx.spring_layout(G)

# draw the nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=500)

# draw the edges
nx.draw_networkx_edges(G, pos, edge_color='gray')

# add labels to the nodes
nx.draw_networkx_labels(G, pos, font_color='black')

# edge weights => labels
edge_labels = {(u, v): distances[u-1][v-1] for u, v in G.edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# naming the x and y coordinates + title
plt.xlabel('Location')
plt.ylabel('Location')
plt.title('Graph Representation of Locations')

# graph display
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# set the output routes(based on the julia script results)
output_routes = [(1, 2), (2, 3), (3, 5), (4, 1), (5, 6), (6, 4)]

G = nx.Graph()

# add nodes to the graph
for i in range(num_locations):
    G.add_node(i + 1)

# add edges to the graph with distances as edge weights
for i in range(num_locations):
    for j in range(i + 1, num_locations):
        distance = distances[i][j]
        G.add_edge(i + 1, j + 1, weight=distance)

# create a layout for the graph
pos = nx.spring_layout(G)

# draw the nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=500)

# draw the edges
nx.draw_networkx_edges(G, pos, edge_color='gray')

# add labels to the nodes
nx.draw_networkx_labels(G, pos, font_color='black')

# add edge weights as labels
edge_labels = {(u, v): distances[u-1][v-1] for u, v in G.edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# draw the output routes using different colors
for route in output_routes:
    nx.draw_networkx_edges(G, pos, edgelist=[route], edge_color='red', width=2.0)

plt.xlabel('Location')
plt.ylabel('Location')
plt.title('Graph Representation of Locations with Output Routes')

# Display the graph
plt.axis('off')
plt.tight_layout()
plt.show()
```
